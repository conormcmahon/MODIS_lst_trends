Next, set up some functions for basic processing of MODIS data

```python

# *************************************************************
# Load a MODIS image
#    Satellite - pick one of 'Terra' or 'Aqua'
#    Time - pick one of 'Day' or 'Night'
def getMODISImage(satellite, time, target_date):
    # ****** Get ImageCollection Path ******
    #   Set ImageCollection path based on user parameters...
    appropriate_satellite_selection = False
    if((satellite == 'Terra')):
        if((time == 'Day')):
            lst_collection_path = "MODIS/061/MOD21A1D"
            appropriate_satellite_selection = True
        if((time == 'Night')):
            lst_collection_path = "MODIS/061/MOD21A1N"
            appropriate_satellite_selection = True
        ndvi_collection_path = "MODIS/061/MOD09GQ"
    if((satellite == 'Aqua')):
        if((time == 'Day')):
            lst_collection_path = "MODIS/061/MYD21A1D"
            appropriate_satellite_selection = True
        if((time == 'Night')):
            lst_collection_path = "MODIS/061/MYD21A1N"
            appropriate_satellite_selection = True
        ndvi_collection_path = "MODIS/061/MYD09GQ"
    if(not appropriate_satellite_selection):
        raise ValueError("In getCollectionMODIS(): Invalid input provided for satellite. Expected \'Terra\' or \'Aqua\' but received \'", satellite, "\'")
    
    # ****** Load imagery ******
    # Get LST and QC (cloud mask) data for target date
    lst_col = ee.ImageCollection(lst_collection_path).filterDate(target_date, target_date.advance(0.99, 'days'))
    lst = lst_col.select('LST_1KM', 'QC').first()
    # Get NDVI data from nearest value in 16-day product
    surf_reflectance_col = ee.ImageCollection(ndvi_collection_path).filterDate(target_date, target_date.advance(0.99, 'days'))
    #surf_reflectance_col = ee.ImageCollection("MODIS/061/MOD13A2").filterDate(target_date.advance(-8, 'days'), target_date.advance(8, 'days'))
    surf_reflectance = surf_reflectance_col.select('sur_refl_b01', 'sur_refl_b02').first()
    # Get Albedo data also from MODIS
    albedo_col = ee.ImageCollection("MODIS/061/MCD43A3").filterDate(target_date, target_date.advance(0.99, 'days'))
    albedo = albedo_col.select('Albedo_WSA_shortwave').first()
    
    # ****** Build output image ******
    # Combine LST and surface reflectance
    output_data = lst.addBands(surf_reflectance).addBands(albedo)
    # Add a band for day-of-year
    output_data = output_data.addBands(ee.Image(target_date.getRelative('day','year')).rename('day_of_year').int())
    # Add a band for year
    output_data = output_data.addBands(ee.Image(target_date.difference(ee.Date('0000-01-01'),'years')).rename('year').int())
    
    # ****** Safety-build of Dummy Image ******
    # Earth Engine behaves a bit strangely if there are NO good images in the target window
    # In those cases, build a dummy image which is masked everywhere, with the property 'retain' set to FALSE
    # Otherwise, copy properties from the source LST image into the output image
    output_data = ee.Algorithms.If((lst_col.size().gte(1)).And(surf_reflectance_col.size().gte(1)),
                                   output_data.set('retain', True).copyProperties(**{'source': lst, 'properties': ['system:time_start']}),
                                   ee.Image.constant(ee.List([0,0,0,0,0,0])).updateMask(ee.Image(0)).set('retain', False))
    return ee.Image(output_data)


# *************************************************************
# Process MODIS imagery - add NDVI and mask for clouds
def processMODISImage(img):
    # Add NDVI band to image
    img_with_ndvi = img.addBands(img.normalizedDifference(['sur_refl_b02', 'sur_refl_b01']).rename("NDVI"))
    # The QA band includes quality information on cloud cover
    QC = img.select('QC').int()
    # In order to be clear, bit 4 should be 0
    cloud_mask = QC.bitwiseAnd(1 >> 4)
    # Update input image to mask cloudy areas 
    return img_with_ndvi.updateMask(cloud_mask.eq(0))


# *************************************************************
# Get a whole LST + NDVI image collection, cloud-masked, for a specified timeseries
# This version creates collection based on start and end date
def getMODISCollection(satellite, time, start_date, end_date):
    # Number of days within specified time period
    num_days = end_date.difference(start_date, 'days').floor()
    # List from 0 to (n-1) days
    day_offsets = ee.List.sequence(0, num_days, 1)
    # Function to get days from offset list
    def getOffsetDate(offset):
        return start_date.advance(offset, 'days')
    # All days within range, as ee.List of ee.Date
    target_dates = day_offsets.map(getOffsetDate)
    # Get collection over each day in the period
    def buildCollection(target_date):
        return ee.Image(getMODISImage(satellite, time, ee.Date(target_date)))
    output_col = ee.ImageCollection(target_dates.map(buildCollection))
    # Filter to remove empty images 
    output_col = output_col.filter(ee.Filter.eq('retain', True))
    # Apply NDVI calculation and cloud mask
    return output_col.map(processMODISImage)


# *************************************************************
# Get a whole LST + NDVI image collection, cloud-masked, for a specified timeseries
# This version based on a specific day of year target
def getMODISCollectionForDOY(satellite, time, doy):
    # Start date
    start_date = ee.Date("2000-01-01").advance(doy-1, 'days')
    # List with years after 2000
    year_offsets = ee.List.sequence(0,24)
    # Function to get days from offset list
    def getOffsetYear(offset):
        return start_date.advance(offset, 'years')
    # All dates within range, as ee.List of ee.Date
    target_dates = year_offsets.map(getOffsetYear)
    # Get collection over each day in the period
    def buildCollection(target_date):
        return ee.Image(getMODISImage(satellite, time, ee.Date(target_date)))
    output_col = ee.ImageCollection(target_dates.map(buildCollection))
    # Filter to remove empty images 
    output_col = output_col.filter(ee.Filter.eq('retain', True))
    # Apply NDVI calculation and cloud mask
    return output_col.map(processMODISImage)


# *************************************************************
# Update NDVI and Albedo bands to use same resolution as LST (via mean aggregation)
def aggregateToLSTResolution(img):
    return img.reduceResolution(reducer=ee.Reducer.mean(), maxPixels=10000, bestEffort=True).reproject(img.select('LST_1KM').projection())
```

```python

# US Cropland Dataset
crop_dataset = ee.ImageCollection('USDA/NASS/CDL').filterDate(ee.Date("2000-01-01"), ee.Date("2023-12-31"));
''' Interpreting crop classes:
   https://developers.google.com/earth-engine/datasets/catalog/USDA_NASS_CDL#bands
1-60,205-209,213-214,216,219,221-222,224-241,243-254 - herbaceous row crops
61 - fallow crop field
63-65,87,131-195 - natural
66-77,204,210-212,215,217-218,220,223,242 - woody crops
82, 121-124 - developed
83, 111 - water
88 - nonag/undefined
92 - aquaculture
112 - snow/ice
'''
def getCoarseClass(img):
    img = img.select('cropland')
    herbaceous_crop = (img.lte(60)
                           .add(img.gte(205).multiply(img.lte(209)))
                           .add(img.gte(213).multiply(img.lte(214)))
                           .add(img.eq(216))
                           .add(img.eq(219))
                           .add(img.gte(221).multiply(img.lte(222)))
                           .add(img.gte(224).multiply(img.lte(241)))
                           .add(img.gte(243).multiply(img.lte(254))))
    woody_crop = ((img.gte(66).multiply(img.lte(77)))
                       .add(img.eq(204))
                       .add(img.gte(210).multiply(img.lte(212)))
                       .add(img.eq(215))
                       .add(img.gte(217).multiply(img.lte(218)))
                       .add(img.eq(220))
                       .add(img.eq(223))
                       .add(img.eq(242)))
    fallow_field = img.eq(61)
    urban_area = (img.eq(82)             # 'Developed
                      .add(img.eq(122))  # 'Developed / Low Intensity'
                      .add(img.eq(123))  # 'Developed / Med Intensity'
                      .add(img.eq(124))) # 'Developed / High Intensity
    natural_land = ((img.gte(63).multiply(img.lte(65)))
                         .add(img.eq(87))
                         .add(img.eq(121)) # 'Developed / Open Space'
                         .add(img.gte(131).multiply(img.lte(195))))
    water = img.eq(83).add(img.eq(111))
    other_cover = (img.eq(88)).add(img.eq(92)).add(img.eq(112))
    return (herbaceous_crop.multiply(1)#
           .add(woody_crop.multiply(2))#
           .add(fallow_field.multiply(3))#
           .add(urban_area.multiply(4))#
           .add(natural_land.multiply(5))#
           .add(water.multiply(6))#
           .add(other_cover.multiply(7)))

coarse_cover_types = crop_dataset.map(getCoarseClass)
coarse_cover_mode = coarse_cover_types.mode()
cover_projection = crop_dataset.select('cropland').filterDate(ee.Date("2020-01-01"), ee.Date("2020-12-31")).filterBounds(ee.Geometry.Point([-115.19, 33.36])).first().projection();
def updateProj(img):
    return img.setDefaultProjection(cover_projection)
coarse_cover_types.map(updateProj)
coarse_cover_mode = coarse_cover_mode.setDefaultProjection(cover_projection)

# Specify some target areas 
imperial_valley_extent = ee.FeatureCollection("projects/corvus-phenology-drylands/assets/southern_california_modis_extent")
smaller_test_area = ee.Geometry.Rectangle([-116.5,32.5,-115,34])
#smaller_test_area = ee.Geometry.Rectangle([-115.75,32.5,-115,33.0])   # another even smaller test version
#smaller_test_area = ee.Geometry.Rectangle([-115.75,33.0,-115.5,33.25])   # another even SMALLER test version

# Save the modal crop data to an Earth Engine asset at MODIS resolution
modis_projection =  getMODISCollection('Aqua', 'Day', ee.Date("2024-01-01"), ee.Date("2024-12-31")).first().select('LST_1KM').projection()
display('MODIS Projection is ', modis_projection)
# Reproject the crop data
coarse_cover_reproj = coarse_cover_mode

# Grid to cover export geometry in tiled form
land_cover_bounds = smaller_test_area.bounds(**{
    'proj': modis_projection, 'maxError': 1
})
land_cover_grid = land_cover_bounds.coveringGrid(**{
    'proj': modis_projection, 'scale': 45*1000
})

tile_ids = land_cover_grid.aggregate_array('system:index').getInfo();
print('Total tiles', len(tile_ids))

def exportCropImage(crop_mask, suffix):
    # Export each tile
    for i, tile_id in enumerate(tile_ids):
        # Get the geometry of the ith tile
        grid_i = ee.Feature(land_cover_grid.toList(1, i).get(0))
        geometry_i = grid_i.geometry()
        task_name = 'tile_' + tile_id.replace(',', '_')
        # Skip cases where the tile doesn't overlap the target geometry
        valid_tile = geometry_i.intersects(smaller_test_area, maxError=10)
        if(not valid_tile.getInfo()):
            print('Skipping task: ', i+1, ' with name ', task_name)
        else:
            task = ee.batch.Export.image.toAsset(**{
                'image': crop_mask.reduceResolution(reducer=ee.Reducer.mean(), maxPixels=10000, bestEffort=True).reproject(modis_projection),
                'description': f'imperial_valley_land_cover_{task_name}_{suffix}',
                'assetId': f'projects/corvus-phenology-drylands/assets/imperial_valley_land_cover_{task_name}_{suffix}',
                #'fileNamePrefix': task_name,
                #'folder':'earth_engine_outputs',
                'crs': modis_projection.getInfo()['crs'],
                'crsTransform': modis_projection.getInfo()['transform'],
                'region': geometry_i,
                'maxPixels': 1e10
            })
            task.start()
            print('Started Task: ', i+1, ' with name ', task_name)
# Do the above for each crop type
#exportCropImage(((coarse_cover_mode.eq(1)).add(coarse_cover_mode.eq(2)).add(coarse_cover_mode.eq(3))).gt(0), "any_crop")
#exportCropImage(coarse_cover_mode.eq(1), "herbaceous_crops")
#exportCropImage(coarse_cover_mode.eq(2), "woody_crops")
#exportCropImage(coarse_cover_mode.eq(3), "fallow_crops")
#exportCropImage(coarse_cover_mode.eq(4), "urban")
exportCropImage(coarse_cover_mode.eq(5), "natural_land")
#exportCropImage(coarse_cover_mode.eq(6), "water")
#exportCropImage(coarse_cover_mode.eq(7), "other")


# Write modal land cover type to an asset at MODIS resolution
task = ee.batch.Export.image.toAsset(
    image = coarse_cover_reproj,
    description = 'imperial_valley_land_cover',
    assetId = 'projects/corvus-phenology-drylands/assets/imperial_valley_land_cover',  # <> modify these
    region = smaller_test_area,
    crs = modis_projection.getInfo()['crs'],
    crsTransform = modis_projection.getInfo()['transform'],
    maxPixels = 1e10
)
#task.start()
```

Now, let's build some methods to compute trends and linear relationships among variables...

```python
# Get the linear relationship (using Mann Kendall non-parametric test) between two variables in an image, aggregated spatially within a region
#   Uses the Sens slope, which should be less sensitive to outliers
def spatialRegression(img, var_1, var_2, region):
    return img.select([var_1, var_2]).reduceRegion(
        reducer = ee.Reducer.sensSlope(),
        geometry = region,
        scale = 1000,
        maxPixels = 1e9)

coarse_cover_loaded = ee.Image('projects/corvus-phenology-drylands/assets/imperial_valley_land_cover_tile_204_89_any_crop')
Map = geemap.Map(center=[33,-115],zoom=6)
Map.add_basemap("Esri.WorldImagery")
Map.addLayer(coarse_cover_mode, {'min': 0, 'max': 7}, "first test")
Map.addLayer(coarse_cover_loaded, {'min': 0, 'max': 1}, "loaded test")
#Map.addLayer(land_cover_grid, {}, "grid")
#Map.addLayer(smaller_test_area, {}, "target_area")
Map
```

```python
## # Spatial extents for project - one larger, one smaller

southern_california_area = ee.FeatureCollection("projects/corvus-phenology-drylands/assets/southern_california_modis_extent")   # larger
imperial_valley_area = ee.Geometry.Rectangle([-116.5,32.5,-115,34])                                                             # smaller

# Next try mapping over collection but returning collection with reducer output set as property?
# Then, iterate again over blocks of collection, one for each day, getting mean values or something?? 
# I think if you map over day, then for each day get subset of collection on that day, then return all collections, and at end sort by time? should work

def buildNormalizedCollection(doy, target_area, sensor, time, mask, min_pixels=100):
    
    # Get all imagery from the target sensor, day-of-year, and time-of-day
    full_collection = getMODISCollectionForDOY('Aqua', 'Day', doy)

    # Apply land cover mask 
    def applyMask(img):
        return img.updateMask(mask)
    full_collection = full_collection.map(applyMask)

    # Count how many valid (unmasked) pixels there are in the dataset
    def countGoodPixels(img):
        # Get count of valid pixels within the target area
        valid_pixel_count = img.select(["NDVI", "LST_1KM"]).reduceRegion(
            reducer = ee.Reducer.count(),
            geometry = imperial_valley_area,
            scale = 1000,
            maxPixels = 1e9) 
        return ee.Image((img.set('valid_lst_count', valid_pixel_count.get('LST_1KM'))
                            .set('valid_ndvi_count', valid_pixel_count.get('NDVI'))))
    full_collection = ee.FeatureCollection(full_collection.map(countGoodPixels))
    # Filter to remove images (years) with NO good imagery (entirely masked, e.g. by clouds)
    #   note - if this isn't set, the aggregation will fail in cases with no good imagery because aggregate_mean() can't handle NaN values for slope/offset
    valid_year_filter = ee.Filter.And(ee.Filter.gte('valid_lst_count', min_pixels), (ee.Filter.gte('valid_ndvi_count', min_pixels)))
    full_collection = ee.ImageCollection(full_collection.filter(valid_year_filter))

    # Get slope and variance for each image
    def getVegTempModel(img):
        # Get slope of relationship between NDVI and LST using Sen's Slope estimator
        regression_result = img.select(["NDVI", "LST_1KM"]).reduceRegion(
            reducer = ee.Reducer.sensSlope(),
            geometry = imperial_valley_area,
            scale = 1000,
            maxPixels = 1e9) 
        # Get variance of LST before and after normalization
        img = img.addBands((img.select('LST_1KM')
                               .subtract(img.select('NDVI')
                                            .multiply(ee.Image(regression_result.getNumber('slope')))))
                               .rename("LST_norm"))
        variance_result = img.select(["LST_1KM", "LST_norm"]).reduceRegion(
            reducer = ee.Reducer.variance(),
            geometry = imperial_valley_area,
            scale = 1000,
            maxPixels = 1e9) 
        # Return a Feature with the regression data, variances, and count of valid pixels
        return (ee.Feature(ee.Geometry(imperial_valley_area).centroid())
                  .set('year', img.date().get('year'))
                  .set('doy', img.date().getRelative('day', 'year'))
                  .set('ndvi_lst_slope', regression_result.get('slope'))
                  .set('ndvi_lst_offset', regression_result.get('offset'))
                  .set('lst_variance', variance_result.get('LST_1KM'))
                  .set('lst_norm_variance', variance_result.get('LST_norm'))
                  .set('valid_lst_count', img.get('valid_lst_count'))
                  .set('valid_ndvi_count', img.get('valid_ndvi_count')))
    
    slope_collection = full_collection.map(getVegTempModel)
    
    average_slope = slope_collection.aggregate_mean('ndvi_lst_slope')
    average_offset = slope_collection.aggregate_mean('ndvi_lst_offset')
    average_lst_variance = slope_collection.aggregate_mean('lst_variance')
    average_lst_norm_variance = slope_collection.aggregate_mean('lst_norm_variance')
    def addAverageValues(img):
        return (img.set('ndvi_lst_slope_avg', average_slope)
                   .set('ndvi_lst_offset_avg', average_offset)
                   .set('lst_variance_avg', average_lst_variance)
                   .set('lst_norm_variance_avg', average_lst_norm_variance))
    slope_collection = slope_collection.map(addAverageValues)

    return slope_collection

    # Map over collection
    # Get slope for each image
    # Add it as a property to the image
    # Aggregate over collection property --> overall median slope
    # Normalize all images in collection by that value 

print('starting...')

def processAllSlopes(sensor, time, mask, name_suffix, min_pixels=100):
    for target_doy in range(1,366):
        slope_collection = buildNormalizedCollection(target_doy, imperial_valley_area, 'Aqua', 'Day', mask, min_pixels)
        print('Working on DOY ', target_doy)
        # Uncomment display for debugging
        #display(col_with_slopes)
        task = ee.batch.Export.table.toAsset(
                collection = slope_collection,
                description = "modis_lst_ndvi_regression_doy_" + str(target_doy) + "_" + name_suffix,
                assetId = "projects/corvus-phenology-drylands/assets/modis_imperial_valley/Aqua_day_modis_lst_doy_" + str(target_doy) + "_" + name_suffix,
                folder = "earth_engine_outputs",
                fileFormat = "CSV")
        task.start()

processAllSlopes('Aqua', 'Day', any_crop.gte(0.8), "aqua_day_crops", 100)
processAllSlopes('Aqua', 'Day', urban.gte(0.8), "aqua_day_urban", 50)
processAllSlopes('Aqua', 'Day', natural_land.gte(0.8), "aqua_day_natural", 100)
```
